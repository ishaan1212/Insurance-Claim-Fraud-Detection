Perform chi-squared test of independence to see whether a particular categorical variable is related with the fraud

```python
from scipy.stats import chi2_contingency
```

```python
significant_features = []
for col in cat_col:

    contingency_table = pd.crosstab(data[col], data['fraud_reported'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    if p < 0.05:
        significant_features.append(col)
        print(f"Chi-squared test result for {col}: p-value = {p}")
    else:
        print("not significant")
```

```python
print("Significant features:", significant_features)
```

Data Cleaning and Preprocessing Starts

Detect and handle the outliers

Using IQR method (interquartile range)

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # 25th percentile
    Q3 = data[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f"{column}:")
    print(f"Lower Bound = {lower_bound}, Upper Bound = {upper_bound}")
    print(f"Number of outliers = {len(outliers)}\n")
    return lower_bound, upper_bound
```

```python
for col in num_col:
    detect_outliers_iqr(data, col)
```

Now we will start handling the outliers

Techniques include capping, removal, and using robust scaling   

We will cap the outlier (winsorization)

```python
for col in num_col:
    lower_bound, upper_bound = detect_outliers_iqr(data, col)
    data[col] = data[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
```

```python
for col in num_col:
    detect_outliers_iqr(data, col)
```

```python
for col in num_col:
    sns.boxplot(data=data, x=col)
    plt.show()
```

Since there are no whiskers so, the outliers have been capped successfully.

Now, we will encode the categorical variables using techniques like Label Encoding, One Hot encoding, Frequency Encoding

ONE HOT ENCODING

```python
# Apply One-Hot Encoding
one_hot_columns = ['policy_state', 'insured_sex', 'insured_hobbies', 'insured_relationship', 
                   'incident_type', 'collision_type', 'authorities_contacted', 'incident_state', 'auto_make']
data = pd.get_dummies(data, columns=one_hot_columns, drop_first=True)
```

LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding
label_columns = ['policy_csl', 'insured_education_level', 'incident_severity', 
                 'property_damage', 'police_report_available']
le = LabelEncoder()

for col in label_columns:
    data[col] = le.fit_transform(data[col])
```

FREQUENCY ENCODING

```python
# Frequency Encoding for High Cardinality Features
high_cardinality_columns = ['insured_occupation', 'incident_city', 'auto_model']

for col in high_cardinality_columns:
    freq_map = data[col].value_counts().to_dict()
    data[col] = data[col].map(freq_map)
```

Still the recall for Y (fraud) is quite low so now we will focus on hypertuning the parameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

You can see above after applying gridsearchcv, recall increased further to 47 and the no fraud recall decreased to 85
